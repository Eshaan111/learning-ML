```python
import numpy as np

class LinearRegressionClosed :

  def __init__(self):
    self.coef_ = None
    self.intercept = 0.0
    pass

  def fit(self,X,Y):
    X = np.array(X)
    Y = np.array(Y)

    # Weight matrix Formula = inv(Xt.x)  * Xt.y

    feature_matrix = np.c_[np.ones((X.shape[0],1)),X]
    fM = feature_matrix
    weight_Matrix = np.linalg.inv(fM.T @ fM) @ (fM.T @ Y)
    wM = weight_Matrix

    self.intercept = wM[0]
    self.coef_ = wM[1:]

  def predict(self,X):
    X = np.array(X)

    return (X @ self.coef_) + self.intercept

```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegressionClosed()
X, y = fetch_california_housing(return_X_y=True)

X_train,X_test,Y_train,Y_test = train_test_split(X,y, test_size = 0.2, random_state=2)
reg.fit(X_train, Y_train)
```

```python
reg_preds = reg.predict(X_test)
```
